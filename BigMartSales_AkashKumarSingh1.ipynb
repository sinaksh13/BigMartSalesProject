```python
# get features and corresponding outcomes
feature_names = ['Product_ID', 'Weight', 'Fat', 'Item_Visibility','Item_Type','Item_MRP','Outlet_Identifier','Outlet_Establishment_Year','Outlet_Size','Outlet_Location_Type','Outlet_Type']
training_features = new_data[feature_names]

outcome_name = ['Item_Outlet_Sales']
outcome_labels = new_data[outcome_name]
```

```python
# view outcome labels
outcome_labels
```

```python
numeric_feature_names = ['Weight', 'Item_Visibility','Item_MRP','Outlet_Establishment_Year']
categoricial_feature_names = ['Product_ID','Fat','Item_Type','Outlet_Identifier','Outlet_Size','Outlet_Location_Type','Outlet_Type']
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

# fit scaler on numeric features
ss.fit(training_features[numeric_feature_names])

# scale numeric features now
training_features[numeric_feature_names] = ss.transform(training_features[numeric_feature_names])

# view updated featureset
training_features
```

```python
training_features = pd.get_dummies(training_features, columns=categoricial_feature_names)
# view newly engineering features
training_features
```

```python
categorical_engineered_features = list(set(training_features.columns) - set(numeric_feature_names))
```
